**Numpy array vs python lists**

```python
#speed
#list
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]

c = []
import time

start = time.time()
for i in range(len(a)):
  c.append(a[i] + b[i])
print(time.time()-start)
```

```python
#speed
#numpy
import numpy as np
a = np.arange(10000000)
b = np.arange(10000000,20000000)

start = time.time()
c = a + b
print(time.time()-start)
```

```python
#memory
#list
a = [i for i in range(10000000)]
import sys

sys.getsizeof(a)
```

```python
a = np.arange(10000000,dtype=np.int8)
sys.getsizeof(a)
```

**Advanced Indexing**

```python
#Normal Indexing and slicing

a = np.arange(24).reshape(6,4)
a
```

```python
#Fancing Indexing

print(a)

print(a[[0,2,3]])

print(a[:,[0,2,3]])
```

```python
#Boolean Indexing
a = np.random.randint(1,100,24).reshape(6,4)
a
```

```python
# find all numbers greater then 50
# a > 50
a[a > 50]
```

```python
# Find out even numbers
# a % 2 == 0
a[a % 2 == 0]
```

```python
# find all numbers greater then 50 and even
# in case of boolean indexing we using bitwise opr

# (a > 50 ) & (a % 2 == 0)
a[(a > 50) & (a % 2 == 0)]
```

```python
# find all number not divisible by 7
# not is ~
# a % 7 == 0

# not divisible
a[~(a % 7 ==0)]

# divisible
a[(a % 7 ==0)]
```

**Broadcasting**

The term broadcasting describes how Numpy treats array with different shapes during arithmetic operations.

The smaller array is "broadcast" across the larger array so that they have compatible shapes

image.png

image.png

```python
# More examples
# the last will be same 3 == 3

a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)
print(b)

print(a+b)
```

```python
 # More examples
# the last will be same 4 == 3 its through a erro

a = np.arange(12).reshape(3,4)
b = np.arange(3)

print(a)
print(b)

print(a+b)
```

```python
# example
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(16).reshape(4,4)
b = np.arange(4).reshape(2,2)

print(a)
print(b)

print(a+b)
```

```python
#sigmoid
def sigmoid(array):
  return 1/(1 + np.exp(-(array)))

a = np.arange(100)
sigmoid(a)
```

```python
# mean squared error -> lose function | linear regression
# means to increase or decrease something the result is coming good

actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)

def mse(actual,predicted):
  return np.mean((actual-predicted)**2)

mse(actual,predicted)
```

**Plotting Graphs**

```python
# Plotting a 2D plot
# x = y
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y = x

plt.plot(x,y)
plt.show()
```

```python
# y = x^2 | parabola
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# y = sin(x)
x = np.linspace(-10,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
# y = xlog(x)
x = np.linspace(-10,10,100)

y = x * np.log(x)

plt.plot(x,y)

plt.show()
```

```python
# sigmoid

x = np.linspace(-10,10,100)
y = 1/(1+np.exp(-x))

plt.plot(x,y)
plt.show()

```
