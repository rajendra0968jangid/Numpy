**np.concatenate**

numpy.concatenate() function concatenate a sequence of array along as existing axis.

```python
#code
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)
```

```python
np.concatenate((c,d),axis=0)

np.concatenate((c,d),axis=1)
```

**np.expand_dims**

With the help of Numpy.expand_dims() method, we can get the expanded dimensions of an array.

```python
#code
a
a.shape
```

```python
np.expand_dims(a,axis=0)
# 1 row 15 cols
np.expand_dims(a,axis=0).shape
```

```python
np.expand_dims(a,axis=1)
# 15 row 1 cols
np.expand_dims(a,axis=1).shape
```

**np.where**

The numpy.where() functions return the indexs of elements in an input array where the given condition is satisfied.

```python
a
```

```python
# find all indexe with value greater then 50

np.where(a > 50)

a[np.where(a>50)]
```

```python
# replace all values > 50 with 0

np.where(a>50,0,a)
```

```python
np.where(a%2==0,0,a)
```
