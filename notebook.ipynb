**What is Numpy?**

**Creating Numpy Arrays**

```python
#np.array
import numpy as np
np.array([1,2,3,4])
# a = np.array([1,2,3,4])
# print(type(a))
```

```python
# 2D and 3D
b = np.array([[1,2,3],[4,5,6]])
print(b)
```

```python
c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
```

```python
#dtype
np.array([1,2,3],dtype=float)
# np.array([1,2,3],dtype=bool)
```

```python
#np.arange()
# np.arange(1,11)
np.arange(1,11,2)
```

```python
# with reshape
np.arange(1,11).reshape(2,5)
```

```python
#np.ones and np.zeros
# np.ones -> 1
np.ones((3,4))

#np.zeros -> 0
np.zeros((3,4))
```

```python
#np.random
np.random.random((3,4))
```

```python
#np.linspace
np.linspace(-10,10,11)
```

```python
#np.identity
np.identity(3)
```

**Array Attributes**

```python
a1 = np.arange(10,dtype=np.int32)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
#ndim
# its give the dimension
a3.ndim
```

```python
#shape
# show the shape with in one row there are 10 items | cols
a1.shape

# show the shape with 3 row and 4 cols
a2.shape

# show the shape with [[[1,2],[3,4]],[[5,6],[7,8]]]
a3.shape
a3
```

```python
#size
a3.size
```

```python
#itemsize
a1.itemsize

a3.itemsize
```

```python
#dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```
