**Array Attributes**

```python
a1 = np.arange(10,dtype=np.int32)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
#ndim
# its give the dimension
a3.ndim
```

```python
#shape
# show the shape with in one row there are 10 items | cols
a1.shape

# show the shape with 3 row and 4 cols
a2.shape

# show the shape with [[[1,2],[3,4]],[[5,6],[7,8]]]
a3.shape
a3
```

```python
#size
a3.size
```

```python
#itemsize
a1.itemsize

a3.itemsize
```

```python
#dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

**Array Operations**

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)

a1
```

```python
#scalar operations

#arithmetic
a1 = a1 * 2
```

```python
#relational
# a1 > 5

a1 == 16
```

```python
#vector operations -> means operations between two np array

#arithmetic
a1 * a2
```

```python
#max/min/sum/prod
#   0 -> col and 1 -> row

np.min(a1)
np.max(a1)
np.sum(a1)
np.prod(a1)

np.max(a1,axis=0)
```

```python
#mean/median/std/var
np.mean(a1)
```

```python
#trigonometric functions
np.sin(a1)
```

```python
# dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
# log and exponents
np.exp(a1)
```

```python
#round/floor/ceil
np.round(np.random.random((2,3))*100)
```
